```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./data/data_01.csv')
```

```python
df.info()
```

```python
plt.scatter(df['corrected V (cm3)'], df['absolute P (hPa)'], label='Data')
plt.plot(df['corrected V (cm3)'], df['Ideal pressures tmax (hPa)'], label='Isotherm at $T_{max}$')
plt.plot(df['corrected V (cm3)'], df['Ideal pressures tmin (hPa)'], label='Isotherm at $T_{min}$')
plt.plot([150]*100, np.linspace(1702.789314, 2096.148065, 100), label='Isochore at $V_{min}$')
plt.plot([295.381911]*100, np.linspace(864.7056151, 1064.459935, 100), label='Isochore at $V_{max}$')
plt.xlabel('Corrected volume (cm$^3$)')
plt.ylabel('Absolute pressure (hPa)')
plt.legend()
plt.show()
```
